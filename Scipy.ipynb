# PDF and CDF

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm
```

```python
x = np.linspace(-6,6,1000)
```

```python
fx = norm.pdf(x, loc = 0, scale = 1)#normal dist-->loc=mean,scale = standard devi
```

```python
plt.plot(x,fx);
```

```python
Fx = norm.cdf(x, loc=0, scale=1)
```

```python
plt.plot(x,Fx);
```

```python
logfx = norm.logpdf(x, loc=0, scale=1)
plt.plot(x,logfx)
```

# Convolution

This functionality used for deep learning , signal processing , digital image processing and statisitcs

```python
from PIL import Image
```

```python
!wget -O flower.jpeg https://st4.depositphotos.com/37740150/40341/i/450/depositphotos_403411206-stock-photo-flower-image-nice-flower-image.jpg
```

```python
im = Image.open('flower.jpeg')
```

```python
plt.imshow(im);
```

```python
gray = np.mean(im, axis=2)
```

```python
gray.shape
```

```python
plt.imshow(gray);
```

```python
plt.imshow(gray, cmap = 'gray');
```

```python
x = np.linspace(-6,6,50)
fx = norm.pdf(x, loc=0, scale=1)
```

```python
filt = np.outer(fx,fx) #Gaussian filter used here
```

```python
plt.imshow(filt, cmap='gray');
```

```python
from scipy.signal import convolve2d
```

```python
out = convolve2d(gray,filt)
```

```python
plt.subplot(1,2,1)#rows,cols,position
plt.imshow(gray,cmap='gray');
plt.subplot(1,2,2)
plt.imshow(out, cmap='gray');
```
